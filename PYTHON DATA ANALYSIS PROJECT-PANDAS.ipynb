#**Q11. WHAT IS THE MEAN VALUE OF EACH COLUMN AGAINST EACH WEATHER CONDITION** ?

```python
import pandas as pd

numeric_columns = data.select_dtypes(include=['int', 'float']).columns

# Group the DataFrame by weather condition
grouped = data.groupby('Weather Condition')

# Calculate the mean value of each numeric column within each group
mean_values = grouped[numeric_columns].mean()
mean_values
```
